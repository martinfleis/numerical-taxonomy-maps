```python
import geopandas
import pandas
```

## Prague

```python
buildings = geopandas.read_file("../epb/geometry.gpkg", layer="buildings")
labels = pandas.read_csv("../epb/prg_cluster_labels_10.csv", index_col=0)
```

```python
buildings
```

```python
labels
```

```python
buildings = buildings.merge(labels, on="uID", how="inner")
```

```python
buildings = buildings.to_crs(4326)
buildings[["cluster", "geometry"]].to_file("prg.geojson", driver="GeoJSON")
```

```python
!tippecanoe -z15 \
           --no-tile-compression \
           --output-to-directory=prg/ \
           --drop-densest-as-needed \
           -f \
           prg.geojson
```
